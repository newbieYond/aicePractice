```python
import pandas as pd
import requests
import sqlite3

url_json = "https://jsonplaceholder.typicode.com/todos"
json = requests.get(url_json).json()
dfJson = pd.DataFrame(json)
print(dfJson.head())
dfJson.to_json('todosJson_records.json', orient='records')
dfJson.to_json('todosJson_index.json', orient='index')


conn = sqlite3.connect('todo_db.db')
dfJson.to_sql('todos', conn, if_exists='replace', index=False)
query = "SELECT * from todos"
dfJson_db = pd.read_sql_query(query, conn)
conn.close()
dfJson_db
```
